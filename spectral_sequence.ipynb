```python
from discrete_flow import Cell, CWComplex, compute_Flo_Hom
```

```python
# Define the cells in the CW complex
x = Cell(0, None, "x")
y = Cell(0, None, "y")
z = Cell(1, [x, y], "z")
w = Cell(1, [x, y], "w")
t = Cell(2, [z, w], "t")
b = Cell(2, [z, w], "b")
f = Cell(3, [t, b], "f")
```

```python
# Define the acyclic partial matching corresponding to our discrete Morse function
X = CWComplex([x, y, z, w, t, b, f])
mu = {
    id(y): w, # mu(y) = w
    id(z): b, # mu(z) = b
}

X.plot_lattice()
```

```python
from spectral_sequence import FloSpecSeq
E = FloSpecSeq(X, mu)
```

```python
E.plot_page1()
```

```python
E.plot_page2()
```

```python
## Torus
```

```python
# Define the cells in the CW complex
a = Cell(0, None, "a")
b = Cell(0, None, "b")
c = Cell(0, None, "c")
d = Cell(0, None, "d")
alpha = Cell(1, [a, b], r"$\alpha$")
beta = Cell(1, [a, b], r"$\beta$")
gamma = Cell(1, [c, d], r"$\gamma$")
delta = Cell(1, [c, d], r"$\delta$")
epsilon = Cell(1, [a, c], r"$\varepsilon$")
zeta = Cell(1, [a, c], r"$\zeta$")
eta = Cell(1, [b, d], r"$\eta$")
theta = Cell(1, [b, d], r"$\theta$")
A = Cell(2, [alpha, epsilon, gamma, eta], "A")
B = Cell(2, [beta, epsilon, delta, eta], "B")
C = Cell(2, [alpha, theta, gamma, zeta], "C")
D = Cell(2, [beta, zeta, delta, theta], "D")
```

```python
# Define the acyclic partial matching corresponding to our discrete Morse function
T = CWComplex([a,b,c,d,alpha,beta,gamma,delta,epsilon,zeta,eta,theta,A,B,C,D])
mu2 = {
    id(b): beta,
    id(c): zeta,
    id(d): delta,
    id(eta): B,
    id(gamma): C,
    id(theta): D,
}

T.plot_lattice()
```

```python
E = FloSpecSeq(T, mu2)
```

```python
E.plot_page1()
```

```python
E.plot_page2()
```
